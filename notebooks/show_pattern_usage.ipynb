```python
df = pd.read_csv("data/cover_map.csv")
```

In the cell below, change the `pattern` variable to see where the respective pattern is used in the data.

Column `pattern_instance` shows which flows belong together

```python
pattern = 0
def highlight_sentiment(row):
    if row["pattern"] == pattern:
        return ['background-color: red'] * len(row)
    else:
        return ['background-color: black'] * len(row)

df.style.apply(highlight_sentiment, axis=1)
```
